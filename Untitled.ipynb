```python
import requests
import pandas as pd
import threading
import datetime
import time

def fetch_weather_data(api_urls, interval=600, output_csv='weather_data.csv'):
    global time_weather_df  # Reference the global DataFrame to update it

    # Initialize the DataFrame with columns, if not already initialized
    if time_weather_df.empty:
        time_weather_df = pd.DataFrame(columns=[
            'timestamp', 'station_id', 'latitude', 'longitude',
            'wind_direction', 'wind_speed', 'humidity', 'precipitation', 'air_temperature'
        ])

    while True:
        current_time = datetime.datetime.now()
        
        wind_directions_data = {}
        wind_speeds_data = {}
        humidity_data = {}
        precipitation_data = {}
        air_temp_data = {}
        station_coords = {}

        for data_type, url in api_urls.items():
            response = requests.get(url)
            if response.status_code == 200:
                data = response.json()
                readings = data['items'][0]['readings']
                for reading in readings:
                    station_id = reading['station_id']
                    value = reading['value']
                    # Assign values by data type
                    if data_type == 'wind_direction':
                        wind_directions_data[station_id] = value
                    elif data_type == 'wind_speed':
                        wind_speeds_data[station_id] = value
                    elif data_type == 'humidity':
                        humidity_data[station_id] = value
                    elif data_type == 'precipitation':
                        precipitation_data[station_id] = value
                    elif data_type == 'air_temperature':
                        air_temp_data[station_id] = value

                # Collect station coordinates only once
                if not station_coords:
                    coords = data['metadata']['stations']
                    for coord in coords:
                        station_id = coord['id']
                        latitude = coord['location']['latitude']
                        longitude = coord['location']['longitude']
                        station_coords[station_id] = (latitude, longitude)
            else:
                print(f"Failed to retrieve data from {url}, status code:", response.status_code)

        # Create new records for the DataFrame
        new_records = []
        for station_id in station_coords:
            # Use most recent data if new data is missing for any type
            recent_records = time_weather_df[time_weather_df['station_id'] == station_id]
            if not recent_records.empty:
                recent_data = recent_records.iloc[-1]
                new_record = {
                    'timestamp': current_time,
                    'station_id': station_id,
                    'latitude': station_coords[station_id][0],
                    'longitude': station_coords[station_id][1],
                    'wind_direction': wind_directions_data.get(station_id, recent_data['wind_direction']),
                    'wind_speed': wind_speeds_data.get(station_id, recent_data['wind_speed']),
                    'humidity': humidity_data.get(station_id, recent_data['humidity']),
                    'precipitation': precipitation_data.get(station_id, recent_data['precipitation']),
                    'air_temperature': air_temp_data.get(station_id, recent_data['air_temperature'])
                }
            else:
            # Handle case where no recent data exists
                new_record = {
                    'timestamp': current_time,
                    'station_id': station_id,
                    'latitude': station_coords[station_id][0],
                    'longitude': station_coords[station_id][1],
                    'wind_direction': wind_directions_data.get(station_id),
                    'wind_speed': wind_speeds_data.get(station_id),
                    'humidity': humidity_data.get(station_id),
                    'precipitation': precipitation_data.get(station_id),
                    'air_temperature': air_temp_data.get(station_id)
                }
            new_records.append(new_record)
        
        # Update DataFrame and save to CSV
        new_df = pd.DataFrame(new_records)
        time_weather_df = pd.concat([time_weather_df, new_df], ignore_index=True)
        time_weather_df.to_csv(output_csv, index=False)

        # Sleep until the next interval
        time.sleep(interval)

# Start the data fetching thread
fetch_weather_interval = 600  # every 10 minutes
thread = threading.Thread(target=fetch_weather_data, args=(api_urls, fetch_weather_interval, 'weather_data.csv'))
thread.start()
```

```python
import plotly.express as px

# Load Panel Plotly extension
pn.extension('plotly')

# Load data function
def load_data():
    return pd.read_csv('weather_data.csv')

# Function to create a plot
def create_plot(df, metric):
    fig = px.line(df, x='timestamp', y=metric, title=f'{metric.capitalize()} over Time',
                  labels={'timestamp': 'Timestamp', metric: f"{metric.capitalize()} Units"})
    return fig

# Dropdown and plot display function
def plot_with_dropdown(metric='humidity'):
    df = load_data()
    return create_plot(df, metric)

# Create a dropdown widget
metric_selector = pn.widgets.Select(name='Metric', options=['humidity', 'precipitation', 'air_temperature'], value='humidity')

# Bind the plot update to the dropdown selector
@pn.depends(metric=metric_selector.param.value)
def update_plot(metric):
    return plot_with_dropdown(metric)

# Define the dashboard layout
dashboard = pn.Column(metric_selector, update_plot)

# Serve the Panel dashboard
dashboard.servable()
```

```python
import pandas as pd

def create_weather_dataframe(station_coords, wind_directions_data, wind_speeds_data, humidity_data, precipitation_data,air_temp_data):
    # Initialize an empty list to store the weather data for each station
    weather_data = []
    
    # Assume all dicts have the same station_id as keys
    for station_id in station_coords:
        # Create a dictionary for each station's weather data
        station_data = {
            'station_id': station_id,
            'latitude': station_coords[station_id][0],
            'longitude': station_coords[station_id][1],
            'wind_direction': wind_directions_data.get(station_id, None),
            'wind_speed': wind_speeds_data.get(station_id, None),
            'humidity': humidity_data.get(station_id, None),
            'precipitation': precipitation_data.get(station_id, None),
            'Air_temperature' : air_temp_data.get(station_id, None),

        }
        # Append the station's weather data to the list
        weather_data.append(station_data)
    
    # Convert the list of dictionaries to a pandas DataFrame
    weather_df = pd.DataFrame(weather_data)
    
    return weather_df

# Assuming all data dicts and station_coords are defined and populated with your data
# Create the DataFrame
weather_df = create_weather_dataframe(station_coords, wind_directions_data, wind_speeds_data, humidity_data, precipitation_data,air_temp_data)
```

```python

import pandas as pd
import panel as pn
pn.extension()

# initialize list of lists
data = [['A','January',10],['A','February',7],['A','March',5],['B','January',4],['B','February',8],['B','March',12] ]

# Create the pandas DataFrame
df = pd.DataFrame(data, columns = ['Type', 'Month','Metric'])

#lists creation
type_list=['A','B']

#widget creation
dd_types = pn.widgets.Select(name='Select the type to display report:', options=type_list)

# See https://pandas.pydata.org/pandas-docs/stable/user_guide/style.html
def custom_style(val):
    color = 'red' if val > 5 else 'green'
    return 'color: %s' % color

#filter df function
def display_panel(value):

    table = df[df.Type==value].pivot_table(values='Metric',index='Type',columns='Month')
    styled_table =  table.style.applymap(custom_style)
    return pn.panel(styled_table)

# create relation between widgets
@pn.depends(dd_types)
def get_parameters(value):
    return display_panel(value)


# arrange and correlate widgets and functions
report_area = pn.Row(dd_types, get_parameters)
report_area.show()
```
