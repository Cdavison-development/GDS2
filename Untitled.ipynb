```python
def preprocess_dataframe(df):
    # Handle all-NA columns potentially causing issues in future pandas versions
    return df.dropna(axis=1, how='all')
```

```python
def round_time_to_nearest_minute(dt):
    # Rounds to the nearest minute by adding 30 seconds and then truncating extra seconds
    return (dt + datetime.timedelta(seconds=30)).replace(second=0, microsecond=0)
```

```python
import requests
import pandas as pd
import threading
import datetime
import time

global time_weather_df  
time_weather_df = pd.DataFrame() 
def fetch_weather_data(api_urls, interval=600, output_csv='weather.csv'):
    global time_weather_df  # Reference the global DataFrame to update it

    # Initialize the DataFrame with columns, if not already initialized
    if time_weather_df.empty:
        time_weather_df = pd.DataFrame(columns=[
            'timestamp', 'station_id', 'latitude', 'longitude',
            'wind_direction', 'wind_speed', 'relative_humidity', 'precipitation', 'air_temperature'
        ])

    while True:
        current_time = datetime.datetime.now()
        rounded_time = round_time_to_nearest_minute(current_time)
        
        wind_directions_data = {}
        wind_speeds_data = {}
        humidity_data = {}
        precipitation_data = {}
        air_temp_data = {}
        station_coords = {}

        for data_type, url in api_urls.items():
            response = requests.get(url)
            if response.status_code == 200:
                
                data = response.json()
                #print(f"Data received for {data_type}: {data}")  # Debug print
                readings = data['items'][0]['readings']
                for reading in readings:
                    station_id = reading['station_id']
                    print(station_id)
                    value = reading['value']
                    print(f"Station ID: {station_id}, Data Type: {data_type}, Value: {value}")  # Debug print
                    if data_type == 'wind_direction':
                        wind_directions_data[station_id] = value
                    elif data_type == 'wind_speed':
                        wind_speeds_data[station_id] = value
                    elif data_type == 'relative_humidity':
                        humidity_data[station_id] = value
                        #print(f"Humidity data updated for {station_id}: {humidity_data[station_id]}")  # Debug print
                    elif data_type == 'precipitation':
                        precipitation_data[station_id] = value
                    elif data_type == 'air_temperature':
                        air_temp_data[station_id] = value

                # Collect station coordinates only once
                if not station_coords:
                    coords = data['metadata']['stations']
                    for coord in coords:
                        station_id = coord['id']
                        latitude = coord['location']['latitude']
                        longitude = coord['location']['longitude']
                        station_coords[station_id] = (latitude, longitude)
            else:
                print(f"Failed to retrieve data from {url}, status code:", response.status_code)

        # Create new records for the DataFrame
        new_records = []
        for station_id in station_coords:
            # Use most recent data if new data is missing for any type
            recent_records = time_weather_df[time_weather_df['station_id'] == station_id]
            if not recent_records.empty:
                recent_data = recent_records.iloc[-1]
                new_record = {
                    'timestamp': rounded_time,
                    'station_id': station_id,
                    'latitude': station_coords[station_id][0],
                    'longitude': station_coords[station_id][1],
                    'wind_direction': wind_directions_data.get(station_id, recent_data['wind_direction']),
                    'wind_speed': wind_speeds_data.get(station_id, recent_data['wind_speed']),
                    'relative_humidity': humidity_data.get(station_id, recent_data['relative_humidity']),
                    'precipitation': precipitation_data.get(station_id, recent_data['precipitation']),
                    'air_temperature': air_temp_data.get(station_id, recent_data['air_temperature'])
                }
            else:
            # Handle case where no recent data exists
                new_record = {
                    'timestamp': rounded_time,
                    'station_id': station_id,
                    'latitude': station_coords[station_id][0],
                    'longitude': station_coords[station_id][1],
                    'wind_direction': wind_directions_data.get(station_id),
                    'wind_speed': wind_speeds_data.get(station_id),
                    'relative_humidity': humidity_data.get(station_id),
                    'precipitation': precipitation_data.get(station_id),
                    'air_temperature': air_temp_data.get(station_id)
                }
            new_records.append(new_record)
        
        # Update DataFrame and save to CSV
        new_df = pd.DataFrame(new_records)
        new_df = preprocess_dataframe(new_df)  # Preprocess to handle potential all-NA columns
        time_weather_df = pd.concat([time_weather_df, new_df], ignore_index=True)
        time_weather_df.to_csv(output_csv, index=False)


        # Sleep until the next interval
        time.sleep(interval)

# Start the data fetching thread
#fetch_weather_interval = 600  # every 10 minutes
#thread = threading.Thread(target=fetch_weather_data, args=(api_urls, fetch_weather_interval, 'weather_data.csv'))
#thread.start()
```

```python
import requests
import pandas as pd
import threading
import datetime
import time

global air_quality_df  
try:
    air_quality_df = pd.read_csv('air_quality.csv', parse_dates=['timestamp'])
except FileNotFoundError:
    air_quality_df = pd.DataFrame(columns=[
        'timestamp', 'region', 'latitude', 'longitude', 'PSI', 'PM10', 'PM25', 'O3', 'NO2', 'CO', 'SO2'
    ])

def fetch_pollution_data(api_url, interval=600):
    global air_quality_df
    while True:
        current_time = datetime.datetime.now()
        rounded_time = round_time_to_nearest_minute(current_time)
        response = requests.get(api_url)
        if response.status_code == 200:
            data = response.json()
            process_pollution_data(data, rounded_time)
        else:
            print(f"Failed to retrieve data, status code:", response.status_code)
        time.sleep(interval)  # Sleep for specified interval (e.g., 600 seconds for 10 minutes)

def process_pollution_data(data, rounded_time):
    new_records = []
    region_metadata = data['region_metadata']
    readings = data['items'][0]['readings']

    for region in region_metadata:
        region_name = region['name']
        latitude = region['label_location']['latitude']
        longitude = region['label_location']['longitude']

        # Creating a new record for each pollutant, assuming a structure to include all pollutants
        new_record = {
            'timestamp': rounded_time,
            'region': region_name,
            'latitude': latitude,
            'longitude': longitude,
            'PSI': readings['psi_twenty_four_hourly'].get(region_name, 'No data'),
            'PM10': readings['pm10_twenty_four_hourly'].get(region_name, 'No data'),
            'PM25': readings['pm25_twenty_four_hourly'].get(region_name, 'No data'),
            'O3': readings['o3_eight_hour_max'].get(region_name, 'No data'),
            'NO2': readings['no2_one_hour_max'].get(region_name, 'No data'),
            'CO': readings['co_eight_hour_max'].get(region_name, 'No data'),
            'SO2': readings['so2_twenty_four_hourly'].get(region_name, 'No data')
        }

        new_records.append(new_record)

    update_pollution_df(new_records)

def update_pollution_df(new_records):
    global air_quality_df
    new_df = pd.DataFrame(new_records)
    new_df['timestamp'] = pd.to_datetime(new_df['timestamp'])

    air_quality_df = pd.concat([air_quality_df, new_df], ignore_index=True)
    air_quality_df = air_quality_df.drop_duplicates(subset=['timestamp', 'region'], keep='last').reset_index(drop=True)
    
    air_quality_df.to_csv('air_quality.csv', index=False)

# Set up the API URL and start the thread
#api_url = "https://api.data.gov.sg/v1/environment/psi"  # Example API URL
#pollution_thread = threading.Thread(target=fetch_pollution_data, args=(api_url, 600))
#pollution_thread.start()
```

```python
import pandas as pd
import panel as pn
import plotly.express as px

pn.extension("plotly")

def load_pollution_data(file_path):
    df = pd.read_csv(file_path, parse_dates=['timestamp'], dayfirst=True)
    return df

pollution_data = load_pollution_data('cleaned_air.csv')

# Mapping of simple names to descriptive titles
pollutant_descriptions = {
    'PM10': 'PM10 (g/m)',
    'PM25': 'PM2.5 (g/m)',
    'O3': 'Ozone (g/m)',
    'NO2': 'Nitrogen Dioxide (g/m)',
    'CO': 'Carbon Monoxide (mg/m)',
    'SO2': 'Sulphur Dioxide (g/m)'
}

def create_pollutant_plot(df, pollutant, region):
    filtered_data = df[df['region'] == region]
    # Use the descriptive name in the title
    title = f"{pollutant_descriptions[pollutant]} Levels in {region} Singapore"
    fig = px.line(filtered_data, x='timestamp', y=pollutant, title=title)
    fig.update_layout(
        plot_bgcolor='#000000',
        paper_bgcolor='#000000',
        title_font_color='white',
        font_color='white',
        xaxis_title="Time",
        yaxis_title=f"{pollutant_descriptions[pollutant]}",
        xaxis=dict(color='white'),
        yaxis=dict(color='white'),
        title_x=0.5  # Center the title
    )
    fig.update_traces(line_color='#FFA500')  # Set line color to orange
    return fig

pollutants = ['PM10', 'PM25', 'O3', 'NO2', 'CO', 'SO2']
regions = list(pollution_data['region'].unique())

toggle_group = pn.widgets.ToggleGroup(name='Pollutant Toggle', options=pollutants, behavior='radio')
region_selector = pn.widgets.Select(name='Region', options=regions)

@pn.depends(pollutant=toggle_group.param.value, region=region_selector.param.value)
def update_plot(pollutant, region):
    plot = create_pollutant_plot(pollution_data, pollutant, region)
    return pn.pane.Plotly(plot, height=400, width=600)  # Explicit sizing

layout = pn.Column(
    pn.Row(toggle_group, region_selector),
    update_plot
)

layout.servable()
```
